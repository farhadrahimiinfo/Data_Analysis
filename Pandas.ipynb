```python
wine_reviews.loc[0, 'country']
```

```python
#iloc is conceptually simpler than loc because it ignores the dataset's indices.
#when we use iloc we treat the dataset like a big matrix (a list of lists), one that we have to index into by position.
wine_reviews.loc[:, ['country', 'taster_twitter_handle', 'points']]
```

```python
#Manipulating the index
#The set_index() method can be used to do the job. Here is what happens when we set_index to the title field:

wine_reviews.set_index("title")
```

```python
#Conditional selection
wine_reviews.country == 'Italy'
```

```python
wine_reviews.loc[wine_reviews.country == 'Italy']
```

```python
wine_reviews.loc[(wine_reviews.country == 'Italy') & (wine_reviews.points >= 92)]
```

```python
wine_reviews.loc[(wine_reviews.country == 'Italy') | (wine_reviews.points >= 90)]
```

```python
#The first is isin. isin is lets you select data whose value "is in" a list of values.
# For example, here's how we can use it to select wines only from Italy or France:

wine_reviews.loc[wine_reviews.country.isin(['Italy', 'France'])]
```

```python
#The second is isnull (and its companion notnull). These methods let you highlight values which are (or are not) empty
wine_reviews.loc[wine_reviews.price.notnull()]
```
